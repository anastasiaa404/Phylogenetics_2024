### **2. Tree reconstruction**

```python
!conda create -n iqtree
```

```python
!source activate iqtree && conda install iqtree -c conda-forge -c bioconda
```

```python
!source activate iqtree && iqtree -h
```

```python
!source activate iqtree && iqtree -s /content/aligned.fasta -m TEST -bb 1000 -nt AUTO
```

- Akaike Information Criterion:           TIM2+F+I
- Corrected Akaike Information Criterion: TIM2+F+I
- Bayesian Information Criterion:         **HKY+F+I**
- **Best-fit model: HKY+F+I chosen according to BIC**

### 3. Merge the outgroup

```python
!wget "https://www.ncbi.nlm.nih.gov/sviewer/viewer.fcgi?id=NC_020759.1&db=nuccore&report=fasta&retmode=text" -O Loxodonta_cyclotis_NC_020759.1.fasta
!wget "https://www.ncbi.nlm.nih.gov/sviewer/viewer.fcgi?id=NC_000934.1&db=nuccore&report=fasta&retmode=text" -O Loxodonta_africana_NC_000934.1.fasta
```

```python
!cat Loxodonta_cyclotis_NC_020759.1.fasta Loxodonta_africana_NC_000934.1.fasta > merged_outgroup.fasta
```

```python
!source activate mafft && mafft --add /content/merged_outgroup.fasta --reorder /content/aligned.fasta > /content/extended_alignment.fasta
```

```python
!grep -E "NC_020759.1|NC_000934.1" /content/extended_alignment.fasta
```

```python
!source activate iqtree && iqtree -s /content/extended_alignment.fasta -g /content/aligned.fasta.treefile -m HKY+F+I -o "NC_020759.1,NC_000934.1"
```
